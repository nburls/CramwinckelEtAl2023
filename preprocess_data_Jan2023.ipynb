```python
dir='/shared/CAREER_PWC/DeepMIP-Eocene/User_Model_Database_v1.0/'
mdir=['CESM','COSMOS','GFDL','HadCM3','HadCM3','INMCM','IPSL','MIROC','NorESM']
model=['CESM1.2_CAM5','COSMOS-landveg_r2413','GFDL_CM2.1','HadCM3B_M2.1aN','HadCM3BL_M2.1aN','INM-CM4-8','IPSLCM5A2','MIROC4m','NorESM1_F']

proj=['piControl','deepmip_sens_1xCO2','deepmip_sens_2xCO2','deepmip_stand_3xCO2','deepmip_sens_4xCO2',\
      'deepmip_stand_6xCO2','deepmip_sens_9xCO2']
```

```python
# read land sea mask
var='sftlf'

lon=np.arange(0,360,1);lat=np.arange(-89.5,89.5+1,1);
nlon=len(lon);nlat=len(lat);
matrix=np.zeros((nlat,nlon,len(model),len(proj)))
for i in range(len(mdir)):
    for j in range(len(proj)):

        m=dir+mdir[i]+'/'+model[i]+'/'+proj[j]+'/v1.0/'+model[i]+'-*-'+var+'*'+'.nc'
        m_ts=dir+mdir[i]+'/'+model[i]+'/'+proj[j]+'/v1.0/'+model[i]+'-*-'+'ts'+'*'+'mean.nc'

        try:
            file= !ls {m}
            data=ap.rnc(var,file[0]);
            #print(data)
            #print(file[0])
            file_ts= !ls {m_ts}
            try:
                olon=ap.rnc('lon',file_ts[0]);olat=ap.rnc('lat',file_ts[0]);
            except:
                olon=ap.rnc('longitude',file_ts[0]);olat=ap.rnc('latitude',file_ts[0]);
            if data.ndim == 3:
                data=np.nanmean(data,0);
            if olat[0]>0:
                data=data[::-1,:];olat=olat[::-1]
            if olon[0]<0:
                data,olon=ap.shiftgrid(0, data, olon, start=True, cyclic=360.0)
                
            data=ap.interp(data,olon,olat,lon,lat)
            matrix[:,:,i,j]=data
            #print(model[i]+'_'+proj[j])
        except:
            matrix[:,:,i,j]=np.nan
            #print(file)
sftlf=matrix; del matrix;
```
